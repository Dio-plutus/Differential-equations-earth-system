## Vektorplot

Beispielcodes für komplexere Visualisierungen finden sich z.B. mit 

[Ecosia](https://www.ecosia.org/)

oder auf:

[Stackoverflow](https://stackoverflow.com/questions)

[Matplotlib.org](https://matplotlib.org/gallery/index.html)

```python
# Definition der Koordinaten
X, Y = np.meshgrid(np.arange(0, 2 * np.pi, .2), np.arange(0, 2 * np.pi, .2))

# Definition der Komponenten des Vektorfeldes
U = np.cos(X)
V = np.sin(Y)

# Definiere Größe der Abbildung
plt.figure(figsize=(10,10))

# FÜGE KONTURPLOT DER STREAMFUNKTION HIER HINZU ...

# Plotte Vektorfeld
q = plt.quiver(X, Y, U, V)

# Speichere Abbildung als PDF-Datei 
plt.savefig('vector_field.pdf', bbox_inches='tight')
plt.show()
```
